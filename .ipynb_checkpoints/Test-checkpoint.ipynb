```python
import os

# Ganti path folder dataset sesuai dengan lokasi dataset kamu
val_path = '/data/Sayyidan/Dataset/Yolo/train/images'

# Ganti nama file txt yang kamu mau buat
file_name = '/data/Sayyidan/Dataset/Yolo/train.txt'

# Buka file txt untuk ditulis
with open(file_name, 'w') as file:
    # Loop semua file di dalam folder dataset
    for root, dirs, files in os.walk(val_path):
        for filename in files:
            # Cek ekstensi file, kalo mau spesifik jenis file tertentu bisa tambahin if disini
            if filename.endswith('.jpg') or filename.endswith('.png'):
                # Tulis path file ke dalam file txt
                file.write(os.path.join(root, filename) + '\n')

print(f"File '{file_name}' udah dibuat dengan sukses!")
```
